**1. Neuron model**

```python
import numpy as np
import matplotlib.pyplot as plt

# input, weightを定義
x = np.array([0,1,0,1,0])
w = np.array([1,2,3,4,5])

# hを計算
sum = np.sum(x*w)
theta = 10
h = sum - theta
print(h)

# 活性化関数を定義
def f(h):
  a = 1
  out = 1/(1+np.exp(-a*h))
  return out #Q

# yを計算
y = f(h) #Q
print(y)
```

**2. perceptron**

```python
# define input and weight
u_1 = 1
u_2 = 1
w_1 = 1
w_2 = 1

# get h
sum = w_1*u_1 + w_2*u_2 #Q
theta = 1.5
h = sum - theta
print(h)

# define activation function
def f(h):
  if h >= 0:
    output = 1
  elif h < 0:
    output = 0
  return output

# get y
y = f(h)
print(y)
```

**3. Linear regression problem**

```python
num_data = 1000
x = np.random.random(size = num_data)
true_w = -3.4
true_b = 4.2
mu = 0
sigma = 0.2
noise = np.random.normal(mu, sigma, size = num_data)
y = true_w * x + true_b + noise
```

```python
plt.scatter(x, y)
```

```python
# wの偏微分を計算
def get_dw(w, b, y_hat, y):
  dw = 2*np.dot(np.transpose(y_hat-y), x)/1000
  return dw

# bの偏微分を計算
def get_db(w, b, y_hat, y):
  db = 2*np.sum(y_hat - y)/1000
  return db

# 学習率を定義
lr = 0.5 # learning_rate

# パラメータの初期値を定義
w = 4 # true = -3.4
b = 8 # true = +4.2

# 学習
for i in range(1000):
  #順伝播
  y_hat = np.dot(x, w) + b

  #逆伝播
  dw = get_dw(w, b, y_hat, y) 
  db = get_db(w, b, y_hat, y) 

  #パラメータ更新
  w = w - lr * dw #Q
  b = b - lr * db #Q
  
  #結果を出力
  print(i, w, b)
```
